```python
#Formulas y usos de matplotlib

import matplotlib.pyplot as plt
```

```python
#Ejemplo de una linea simple
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 2, 3, 4, 5]

plt.plot(x, y) #Mostrar la grafica
plt.scatter(x, y) #Mostrar puntos en la grafica
plt.xlabel('Eje X') #Etiqueta del eje X
plt.ylabel('Eje Y') #Etiqueta del eje Y
plt.title('Grafica Simple') #Agregar titulo
plt.show() #Mostrar la grafica
```

```python
#Grafico de barras

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values) #Grafico de barras
plt.title("Grafico de barras") #Titulo
plt.xlabel("Categorias") #Etiqueta eje X
plt.ylabel("Valores") #Etiqueta eje Y
plt.show() #Mostrar grafica
```

```python
#Modificaciones de colores, formas, etc
# plt.style.available --->Ver estilos disponibles
x= [0, 1, 2, 3, 4, 5]
y = [0, 1, 2, 3, 4, 5]

plt.plot(x, y, color='blue', marker='o', linestyle='--', linewidth=2, markersize=8)
plt.ylabel("Valores medidos")
plt.xlabel("Tiempo (s)")
plt.title("Grafica con modificaciones")
plt.show()
```
