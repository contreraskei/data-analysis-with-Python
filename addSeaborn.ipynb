```python
#Seaborn libreria de Matplotlib para graficas estadisticas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Exportar base de datos
data = sns.load_dataset("iris")  #Cargar conjunto de datos iris
data #Mostrar datos
```

```python
#Grafica de dispersiÃ³n 
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.xlabel("Longitud del sÃ©palo (cm)")
plt.ylabel("Anchura del sÃ©palo (cm)")
plt.title("DispersiÃ³n de sÃ©palo por especie")
plt.show()
```

```python
#Grafica de Ridgeplot

setosa = data[data['species'] == 'setosa']
versicolor = data[data['species'] == 'versicolor']
virginica = data[data['species'] == 'virginica']

fig, ax = plt.subplots(figsize=(8, 6))
plt.xlabel("Longitud del pÃ©talo (cm)")
#Crear el ridge plot usando kdeplot
sns.kdeplot(data=setosa ["petal_length"], label="setosa", ax = ax, fill= True)
sns.kdeplot(data=versicolor ["petal_length"], label="versicolor", ax = ax, fill= True)
sns.kdeplot(data=virginica ["petal_length"], label="virginica", ax = ax, fill= True)

#Ajustes posiciÃ³n de leyendas
ax.legend(loc="upper right")
plt.title("Ridge plot - Iris dataset")
plt.show()
```

```python
# Grafica de dispersion - personalizada
sns.set_style("whitegrid")
palette = sns.color_palette("husl", 3)

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data, palette=palette)
plt.xlabel("Longitud del sÃ©palo (cm)")
plt.ylabel("Anchura del sÃ©palo (cm)")
plt.title("Grafico personalizado")
plt.show()
```

```python
#GRAFICOS AVANZADOS
#Ejemplo: Supongamos que estamos trabajando con un conjunto de datos sobre ventas de automoviles, 
#que incluye la edad del automovil en aÃ±os y el precio de venta en miles de dolares

#Generar datos de ejemplo
np.random.seed(42) #Para reproducibilidad
edad_autos =np.random.randint (0, 20, size=200) #Edad entre 1 y 20 aÃ±os, 200 automoviles
precio_autos = 30 - edad_autos + np.random.normal(-3, 3, size=200) #Precio base menos depreciacion por edad
data= pd.DataFrame({
    "Edad":edad_autos,
    "Precio":precio_autos
})    

data 
```

```python
#CreaciÃ³n de la grafica avanzada

fig, ax =plt.subplots(2, 2, figsize=(12, 10))

#Grafica de dispersiÃ³n 
ax[0, 0].scatter(data["Edad"], data["Precio"], color='blue', alpha=0.6)
ax[0, 0].set_title("Edad vs Precio de Venta")
ax[0, 0].set_xlabel("Edad del Automovil (aÃ±os)")
ax[0, 0].set_ylabel("Precio de Venta (miles de dolares)")


#Histograma de Edad
sns.histplot(data["Edad"], kde=True, color='green', ax=ax[0, 1])
ax[0, 1].set_title("Histograma de Edad")
ax[0, 1].set_xlabel("Edad del Automovil (aÃ±os)")
ax[0, 1].set_ylabel("Frecuencia")

#Histograma de Precio
sns.histplot(data["Precio"], kde=True, color='r', ax=ax[1, 0])
ax[1, 0].set_title("Histograma de Precio")
ax[1, 0].set_xlabel("Precio de Venta (miles de dolares)")
ax[1, 0].set_ylabel("Frecuencia")

#Eliminar cuarto subplot vacio

ax[1, 1].axis('off')

#Ajuste de espacio entre subplots
plt.subplots_adjust(hspace=0.3, wspace=0.3)

plt.show() #Utiliza esto para visualizar si lo q tienes es un script de python


```
