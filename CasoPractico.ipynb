```python
#Casos practicos Analisis exploratorio
#Utilizando el dataset "tips" de la libreria seaborn, veremos que factores afectan a que las propinas cambien.
#Se investigará como diferentes factores como el dia de la semana, el genero, el tamaño de la mesa, entre otros, afectan a la propina que se deja en un restaurante.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="darkgrid") #Establece el estilo de las graficas.
df= sns.load_dataset("tips") #Carga el dataset tips de la libreria seaborn
df.head() #Muestra las primeras 5 filas del dataframe
```

```python
#Grafico de barras de propinas por sexo

sns.barplot(data=df, x="sex", y="tip", errorbar=None)
plt.title("Propinas promedio por sexo")
plt.xlabel("Sexo")
plt.ylabel("Propina")
plt.show()
# Basados en los resultados, podemos observar que los hombres tienden a dejar propinas ligeramente más altas en promedio en comparación con las mujeres.
```

```python
#Grafico de barras de propinas por dia de la semana
sns.barplot(data=df, x="day", y="tip", errorbar=None)
plt.title("Propinas promedio por día de la semana")
plt.xlabel("Día de la semana")
plt.ylabel("Propina")
plt.show()
#El grafico muestra que el domingo es el dia con mayor propina promedio, seguido del sabado. Los dias entre semana tienen propinas promedio mas bajas.
```

```python
#Grafico de barras de propinas promedio por tiempo
sns.barplot(data=df, x="time", y="tip", errorbar=None)
plt.title("Propinas promedio por tiempo del día")
plt.xlabel("Tiempo del día")
plt.ylabel("Propina")
plt.show()
#El grafico muestra que las propinas promedio son mayores durante la cena en comparación con el almuerzo.
```

```python
#Grafico de barras de propinar si el cliente es fumador o no
sns.barplot(data=df, x="smoker", y="tip", errorbar=None)
plt.title("Propinas promedio según si el cliente es fumador")
plt.xlabel("Fumador")
plt.ylabel("Propina")
plt.show()
#El grafico muestra que los clientes dejan igual propina sean fumadores o no. Es decir, no afecta en nada.
```

```python
#Analisis multivariado
#Grafico de barras de propinas por dia y sexo

sns.barplot(data=df, x="day", y="tip", hue="sex", errorbar=None)
plt.title("Propinas promedio por día y sexo")
plt.xlabel("Día de la semana")
plt.ylabel("Propina")
plt.show()
```

```python
#Grafico que represente las propinas promedios por tiempo y sexo

sns.barplot(data=df, x="time", y="tip", hue="sex", errorbar=None)
plt.title("Propinas promedio por tiempo del día y sexo")
plt.xlabel("Tiempo del día")
plt.ylabel("Propina")
plt.show()
```

```python
#Grafica de propinas promedios por fumador y sexo

sns.barplot(data=df, x="smoker", y="tip", hue="sex", errorbar=None)
plt.title("Propinas promedio según si el cliente es fumador y sexo")
plt.xlabel("Fumador")
plt.ylabel("Propina")
plt.show()
```
