```python
#Ejercicio de Analisis Explotario y Analisis Univariado

#Importar las librerias a utilizar

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set(style="darkgrid")

df = sns.load_dataset("tips")
df 
df.head() #Es util para ver las primeras filas del dataframe
```

```python
df.info() #Nos da informacion sobre el dataframe
df.describe() #Nos da un resumen estadistico del dataframe
```

```python
#Analisis Univariado consiste en analizar una variable a la vez

sns.histplot(data=df, x= "total_bill") #Histograma de la variable total_bill
plt.plot()
```

```python
sns.boxplot(data=df, x="total_bill") #Diagrama de caja de la variable total_bill
plt.plot()
```

```python
sns.kdeplot(data=df, x="total_bill") #Grafica de densidad de la variable total_bill
plt.plot()
```
