```python
#Analis Exploratorio Bivariado y Multivariado

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="darkgrid")

df = sns.load_dataset("tips")
df.head()
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
```

```python
sns.barplot(data=df, x="sex", y="tip")
plt.show()



```

```python
sns.boxplot(data=df, x="sex", y="total_bill")
plt.show()  
```

```python
#ANALISIS MULTIVARIADO
#Hace una comparacin entre todas las variables numricas del dataframe
sns.pairplot(df, hue="sex")
plt.show()

#Esta grafica permite interpretar la correlacin entre las variables numricas del dataframe
```
